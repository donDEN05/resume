```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Комплексное число z = a + bi
a, b = 4, -3
print(f'Длина вектора - {np.sqrt(a ** 2 + b ** 2)}, Угол относитель оси Х - {np.arctan(b/a)}')
# Создание графика
plt.figure()
plt.axhline(0, color='black', linewidth=0.5)  # Действительная ось
plt.axvline(0, color='black', linewidth=0.5)  # Мнимая ось
plt.quiver(0, 0, a, b, angles='xy', scale_units='xy', scale=1, color='blue')  # Вектор
plt.plot(a, b, 'ro')  # Точка (a, b)
plt.text(a, b, f' z = {a} + {b}i', fontsize=12, ha='left')  # Подпись точки

# Настройка графика
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.xlabel('Вещественная')
plt.ylabel('Мнимая часть')
plt.title('Геометрическое представление комплексного числа')
plt.show()
```
