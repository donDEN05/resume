СОЗДАНИЕ СВОДНОЙ ТАБЛИЦЫ

```python
rate = rate[rate.rating > 5.5]
rate = rate[rate.anime_id < 40]
rate = rate[rate.user_id < 13]
new_rate = pd.pivot_table(rate, values='rating', columns='user_id', index='anime_id')
new_rate
```

fillna() заменяет "NaN" на какое то значение

```python
new_rate.fillna(0)
```
